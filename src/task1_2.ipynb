```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("data/task1_1_1.csv", encoding="gbk")
data2 = pd.read_csv("data/task1_1_2.csv", encoding="gbk")
data3 = pd.read_csv("data/task1_1_3.csv", encoding="gbk")
```

```python
# data1和data2建立联系
data1.head()
```

```python
data2.head()
```

```python
data1_2=pd.merge(data1,data2,how="left",left_on='CardNo',right_on='CardNo')
```

```python
data1_2.head()
```

```python
data1_2.apply(lambda x : sum(x.isnull())/len(x), axis=0)
```

```python
missing_rows = data1_2[data1_2.isnull().any(axis=1)] 
missing_rows
```

```python
# 处理缺失值
# 缺失值处理，所选列有空即删除该行
print('删除缺失值前：', data1_2.shape)
data1_2 = data1_2.dropna(subset=['Date'], how='any')
print('删除缺失值后：', data1_2.shape)
```

```python
data1_2.drop_duplicates(subset=['CardNo', 'CardCount'])
data1_2.shape
```

```python
data1_2.apply(lambda x : sum(x.isnull())/len(x), axis=0)
```

```python
data1_2.to_csv('data/task1_2_1.csv', index=False, encoding='gbk')
```

```python
# data1和data3链接
data1_3 = pd.merge(data1, data3, how='left', left_on='AccessCardNo', right_on='AccessCardNo')
```

```python
data1_3.head()
```

```python
data1_3.apply(lambda x : sum(x.isnull())/len(x), axis=0)
```

```python
missing_rows = data1_3[data1_3.isnull().any(axis=1)]
missing_rows
```

```python
data1_3 = data1_3.dropna(subset=['Date'], how='any')
data1_3.to_csv('data/task1_2_2.csv', index=False, encoding='gbk')
```
